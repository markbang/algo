```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from scipy import stats

# 设置中文显示
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 假设数据已加载
# data = pd.read_excel('文献.xlsx')

# 数据探索
print("数据集基本信息:")
print(f"数据条目数: {len(data)}")
print(f"数据列: {', '.join(data.columns)}")
print("\n数据前5行:")
print(data.head())


# 1. 分析普赖斯指数增长定律 - 文献随时间增长趋势
def analyze_price_law(data):
    # 假设数据中有发表年份列，例如"年份"或"发表时间"
    # 根据您的实际数据调整列名
    year_column = "PY"  # 替换为实际年份列名

    if year_column in data.columns:
        # 统计每年的文献数量
        yearly_counts = data[year_column].value_counts().sort_index()
        years = yearly_counts.index.tolist()
        counts = yearly_counts.values

        # 计算累积文献数量
        cumulative_counts = np.cumsum(counts)

        # 可视化
        fig, ax = plt.subplots(figsize=(12, 6))
        ax.bar(years, counts, color="skyblue", alpha=0.7, label="年度文献数")
        ax2 = ax.twinx()
        ax2.plot(years, cumulative_counts, "r-", marker="o", label="累积文献数")

        ax.set_xlabel("年份")
        ax.set_ylabel("年度文献数量")
        ax2.set_ylabel("累积文献数量")
        ax.set_title("文献增长趋势分析 (普赖斯指数增长定律验证)")
        ax.grid(True, alpha=0.3)

        # 合并图例
        lines1, labels1 = ax.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax2.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

        plt.tight_layout()
        plt.show()

        # 拟合指数增长模型
        if len(years) > 5:  # 确保有足够的数据点
            # 转换年份为相对年份（从0开始）
            relative_years = np.array(years) - min(years)

            # 对数变换后进行线性回归
            log_counts = np.log(cumulative_counts)
            slope, intercept, r_value, p_value, std_err = stats.linregress(
                relative_years, log_counts
            )

            # 计算增长率和倍增时间
            growth_rate = slope
            doubling_time = np.log(2) / growth_rate if growth_rate > 0 else float("inf")

            print(f"\n普赖斯定律分析结果:")
            print(f"年增长率: {growth_rate:.4f}")
            print(f"文献倍增时间: {doubling_time:.2f} 年")
            print(f"拟合优度(R²): {r_value**2:.4f}")

            # 可视化拟合结果
            plt.figure(figsize=(10, 6))
            plt.scatter(relative_years, log_counts, color="blue", label="实际数据")
            fit_line = intercept + slope * relative_years
            plt.plot(
                relative_years, fit_line, "r-", label=f"拟合线 (R²={r_value**2:.4f})"
            )
            plt.xlabel("相对年份")
            plt.ylabel("文献累积数量对数值")
            plt.title("普赖斯定律指数增长拟合分析")
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

    else:
        print(f"未找到年份列，请确认数据中包含年份信息并调整列名。")


# 2. 分析布拉德福分散定律 - 期刊分布
def analyze_bradford_law(data):
    # 假设数据中有期刊名称列，例如"期刊"或"来源出版物"
    journal_column = "WE"  # 替换为实际期刊列名

    if journal_column in data.columns:
        # 统计期刊文献数量
        journal_counts = data[journal_column].value_counts()
        total_journals = len(journal_counts)
        total_papers = len(data)

        print(f"\n布拉德福定律分析:")
        print(f"总期刊数: {total_journals}")
        print(f"总文献数: {total_papers}")

        # 按文献数量降序排列期刊
        journal_counts_sorted = journal_counts.sort_values(ascending=False)

        # 计算累积文献占比
        cumulative_papers = np.cumsum(journal_counts_sorted.values)
        cumulative_percentage = 100 * cumulative_papers / total_papers

        # 计算期刊占比
        journal_percentage = (
            100 * np.arange(1, len(journal_counts_sorted) + 1) / total_journals
        )

        # 布拉德福区域划分(尝试划分为3个区域，每个区域约1/3的文献)
        zone_boundaries = []
        target_percentages = [33.33, 66.67]
        for target in target_percentages:
            idx = np.argmax(cumulative_percentage >= target)
            zone_boundaries.append(idx)

        # 计算各区期刊数
        zone_journals = [
            zone_boundaries[0],
            zone_boundaries[1] - zone_boundaries[0],
            len(journal_counts_sorted) - zone_boundaries[1],
        ]

        # 计算各区文献数
        zone_papers = [
            cumulative_papers[zone_boundaries[0] - 1],
            cumulative_papers[zone_boundaries[1] - 1]
            - cumulative_papers[zone_boundaries[0] - 1],
            total_papers - cumulative_papers[zone_boundaries[1] - 1],
        ]

        # 输出各区域信息
        zones = ["核心区", "第二区", "第三区"]
        print("\n布拉德福分区结果:")
        for i, zone in enumerate(zones):
            print(
                f"{zone}: {zone_journals[i]} 种期刊 ({zone_journals[i] / total_journals * 100:.2f}%), "
                f"{zone_papers[i]} 篇文献 ({zone_papers[i] / total_papers * 100:.2f}%)"
            )

        # 可视化布拉德福分布
        plt.figure(figsize=(12, 6))
        plt.plot(journal_percentage, cumulative_percentage, "b-", marker="")

        # 添加各区域分界线
        for i, boundary in enumerate(zone_boundaries):
            boundary_x = 100 * boundary / total_journals
            plt.axvline(x=boundary_x, color="r", linestyle="--")
            plt.text(boundary_x + 1, 50, f"区域{i + 1}-{i + 2}分界", rotation=90)

        plt.xlabel("期刊累积百分比 (%)")
        plt.ylabel("文献累积百分比 (%)")
        plt.title("布拉德福期刊分散定律分析")
        plt.grid(True, alpha=0.3)
        plt.show()

        # 可视化各区域期刊数量
        plt.figure(figsize=(10, 6))
        plt.bar(zones, zone_journals, color=["#3274A1", "#E1812C", "#3A923A"])
        plt.xlabel("布拉德福分区")
        plt.ylabel("期刊数量")
        plt.title("各区域期刊数量分布")
        for i, v in enumerate(zone_journals):
            plt.text(i, v + 0.5, str(v), ha="center")
        plt.show()

    else:
        print(f"未找到期刊列，请确认数据中包含期刊信息并调整列名。")


# 3. 分析洛特卡定律 - 作者发文数量分布
def analyze_lotka_law(data):
    # 假设数据中有作者列，例如"作者"或"第一作者"
    author_column = "AU"  # 替换为实际作者列名

    if author_column in data.columns:
        # 统计每位作者的发文数量
        author_counts = data[author_column].value_counts()

        # 统计发表n篇论文的作者人数
        paper_count_distribution = Counter(author_counts.values)
        max_papers = max(paper_count_distribution.keys())

        paper_counts = list(range(1, max_papers + 1))
        author_counts_list = [paper_count_distribution.get(i, 0) for i in paper_counts]

        # 计算理论分布 (洛特卡定律)
        total_authors = sum(author_counts_list)
        authors_with_one_paper = author_counts_list[0]
        theoretical_counts = [authors_with_one_paper / (i**2) for i in paper_counts]

        print(f"\n洛特卡定律分析:")
        print(f"总作者数: {total_authors}")
        print(f"发表1篇论文的作者数: {authors_with_one_paper}")
        print(
            f"发表≥10篇论文的作者数: {sum([paper_count_distribution.get(i, 0) for i in range(10, max_papers + 1)])}"
        )

        # 可视化
        plt.figure(figsize=(12, 6))
        plt.bar(paper_counts, author_counts_list, alpha=0.7, label="实际分布")
        plt.plot(
            paper_counts, theoretical_counts, "r-", marker="o", label="理论分布 (1/n²)"
        )
        plt.xlabel("论文数量 (n)")
        plt.ylabel("作者数量")
        plt.title("洛特卡科学生产力定律分析")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.xticks(np.arange(0, max_papers + 1, 1 if max_papers < 20 else 2))
        plt.show()

        # 对数坐标可视化
        plt.figure(figsize=(12, 6))
        non_zero_mask = np.array(author_counts_list) > 0
        non_zero_papers = np.array(paper_counts)[non_zero_mask]
        non_zero_authors = np.array(author_counts_list)[non_zero_mask]
        non_zero_theoretical = np.array(theoretical_counts)[non_zero_mask]

        plt.loglog(non_zero_papers, non_zero_authors, "bo", alpha=0.7, label="实际分布")
        plt.loglog(
            non_zero_papers,
            non_zero_theoretical,
            "r-",
            marker="o",
            label="理论分布 (1/n²)",
        )
        plt.xlabel("论文数量 (log)")
        plt.ylabel("作者数量 (log)")
        plt.title("洛特卡定律对数坐标分析")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    else:
        print(f"未找到作者列，请确认数据中包含作者信息并调整列名。")


# 分析文献信息流增长规律
def analyze_information_flow_growth(data):
    """
    分析文献信息流增长规律，包括：
    - 指数增长模式（普赖斯定律）
    - S型增长曲线
    - 线性增长
    """
    # 假设年份列是'PY'
    year_column = "PY"

    if year_column in data.columns:
        # 转换年份为数值类型(如果需要)
        if data[year_column].dtype == "object":
            try:
                data[year_column] = pd.to_numeric(data[year_column])
            except:
                print("无法将年份转换为数值，请检查数据格式")
                return

        # 统计每年的文献数量
        yearly_counts = data[year_column].value_counts().sort_index()
        years = yearly_counts.index.tolist()
        counts = yearly_counts.values

        # 计算累积文献数量
        cumulative_counts = np.cumsum(counts)

        # 可视化年度和累积增长趋势
        fig, ax = plt.subplots(figsize=(14, 8))

        # 年度文献数量柱状图
        ax.bar(years, counts, color="skyblue", alpha=0.7, label="年度文献数量")

        # 设置右侧Y轴显示累积数量
        ax2 = ax.twinx()
        ax2.plot(
            years,
            cumulative_counts,
            "r-",
            marker="o",
            linewidth=2,
            label="累积文献数量",
        )

        ax.set_xlabel("年份", fontsize=12)
        ax.set_ylabel("年度文献数量", fontsize=12)
        ax2.set_ylabel("累积文献数量", fontsize=12)
        ax.set_title("文献信息流增长趋势分析", fontsize=14)
        ax.grid(True, alpha=0.3)

        # 添加数据标签
        for i, count in enumerate(counts):
            ax.text(years[i], count + max(counts) * 0.03, f"{count}", ha="center")

        # 优化X轴刻度显示
        if len(years) > 10:
            step = len(years) // 10
            ax.set_xticks(years[::step])

        # 合并图例
        lines1, labels1 = ax.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax2.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=10)

        plt.tight_layout()
        plt.show()

        # 分析增长模式 - 拟合不同模型
        if len(years) > 5:  # 确保有足够的数据点
            # 转换为相对年份
            years_array = np.array(years)
            relative_years = years_array - years_array.min()

            # 1. 指数增长模型拟合 (普赖斯定律)
            # 对数变换后进行线性回归
            log_counts = np.log(cumulative_counts)
            slope_exp, intercept_exp, r_exp, p_exp, std_exp = stats.linregress(
                relative_years, log_counts
            )

            # 2. 线性增长模型拟合
            slope_lin, intercept_lin, r_lin, p_lin, std_lin = stats.linregress(
                relative_years, cumulative_counts
            )

            # 3. S型曲线(Logistics)拟合
            try:
                from scipy.optimize import curve_fit

                def logistic_func(x, L, k, x0):
                    return L / (1 + np.exp(-k * (x - x0)))

                # 初始参数猜测
                p0 = [max(cumulative_counts) * 1.5, 0.5, np.mean(relative_years)]

                # 尝试进行logistic拟合
                try:
                    popt, pcov = curve_fit(
                        logistic_func, relative_years, cumulative_counts, p0=p0
                    )
                    logistic_fitted = logistic_func(relative_years, *popt)
                    r_logistic = np.corrcoef(cumulative_counts, logistic_fitted)[0, 1]
                    logistic_fit_success = True
                except:
                    logistic_fit_success = False
                    r_logistic = 0
            except ImportError:
                logistic_fit_success = False
                r_logistic = 0

            # 比较模型拟合优度
            r_squared_exp = r_exp**2
            r_squared_lin = r_lin**2
            r_squared_logistic = r_logistic**2

            print("\n信息流增长模型拟合结果:")
            print(f"指数增长模型拟合优度 (R²): {r_squared_exp:.4f}")
            print(f"线性增长模型拟合优度 (R²): {r_squared_lin:.4f}")
            if logistic_fit_success:
                print(f"S型曲线模型拟合优度 (R²): {r_squared_logistic:.4f}")

            # 确定最佳拟合模型
            best_fit = "指数增长"
            best_r2 = r_squared_exp

            if r_squared_lin > best_r2:
                best_fit = "线性增长"
                best_r2 = r_squared_lin

            if logistic_fit_success and r_squared_logistic > best_r2:
                best_fit = "S型曲线(Logistics)"
                best_r2 = r_squared_logistic

            print(f"\n最佳拟合模型: {best_fit} (R² = {best_r2:.4f})")

            if best_fit == "指数增长":
                growth_rate = np.exp(slope_exp) - 1
                doubling_time = np.log(2) / slope_exp if slope_exp > 0 else float("inf")
                print(f"年均增长率: {growth_rate * 100:.2f}%")
                print(f"文献数量倍增时间: {doubling_time:.2f} 年")
                print(
                    "\n结论: 文献呈现普赖斯指数增长规律，符合科学文献早期快速发展的特征。"
                )
            elif best_fit == "线性增长":
                annual_increase = slope_lin
                print(f"年均增加文献数: {annual_increase:.2f} 篇")
                print("\n结论: 文献呈现线性增长规律，表明该领域处于稳定发展阶段。")
            elif best_fit == "S型曲线(Logistics)":
                L, k, x0 = popt
                print(f"预计饱和值(L): {L:.2f}")
                print(f"增长率(k): {k:.4f}")
                print(f"拐点(x0): 相对年份 {x0:.2f} 年")
                print(
                    "\n结论: 文献呈现S型增长规律，表明该领域正经历从快速发展到逐渐饱和的转变过程。"
                )

            # 可视化拟合结果
            plt.figure(figsize=(12, 8))
            plt.scatter(years, cumulative_counts, color="blue", label="实际累积文献数")

            # 生成拟合曲线的预测数据点
            years_range = np.linspace(min(years), max(years), 100)
            relative_years_range = years_range - min(years)

            # 指数模型预测
            exp_fit = np.exp(intercept_exp + slope_exp * (years_range - min(years)))
            plt.plot(
                years_range, exp_fit, "r-", label=f"指数模型 (R²={r_squared_exp:.4f})"
            )

            # 线性模型预测
            lin_fit = intercept_lin + slope_lin * (years_range - min(years))
            plt.plot(
                years_range, lin_fit, "g--", label=f"线性模型 (R²={r_squared_lin:.4f})"
            )

            # S型曲线模型预测(如果成功拟合)
            if logistic_fit_success:
                s_fit = logistic_func(relative_years_range, *popt)
                plt.plot(
                    years_range,
                    s_fit,
                    "m-.",
                    label=f"S型曲线 (R²={r_squared_logistic:.4f})",
                )

            plt.xlabel("年份")
            plt.ylabel("累积文献数量")
            plt.title("文献信息流增长规律拟合分析")
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

            # 计算年增长率
            yearly_growth_rates = []
            for i in range(1, len(years)):
                prev_count = cumulative_counts[i - 1]
                curr_count = cumulative_counts[i]
                growth_rate = (
                    (curr_count - prev_count) / prev_count
                    if prev_count > 0
                    else float("inf")
                )
                yearly_growth_rates.append(growth_rate)

            # 可视化年增长率变化
            plt.figure(figsize=(10, 6))
            plt.bar(years[1:], np.array(yearly_growth_rates) * 100, color="orange")
            plt.axhline(
                y=np.mean(yearly_growth_rates) * 100,
                color="r",
                linestyle="-",
                label=f"平均增长率: {np.mean(yearly_growth_rates) * 100:.2f}%",
            )
            plt.xlabel("年份")
            plt.ylabel("年增长率(%)")
            plt.title("文献年度增长率变化")
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

    else:
        print(f"未找到年份列，请确认数据中包含年份信息并调整列名。")


# 执行分析
print("开始分析文献数据...")
analyze_price_law(data)
analyze_bradford_law(data)
analyze_lotka_law(data)
print("开始分析文献信息流增长规律...")
analyze_information_flow_growth(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 设置中文显示
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False


def analyze_literature_aging(data):
    """分析文献老化：计算老化系数和半衰期"""
    # 假设我们有引用年份列和被引用文献的发表年份列
    data["CR"] = data["CR"].str.extract(r"(\d{4})")  # 提取引用年份
    # 转化为整数
    data["CR"] = pd.to_numeric(data["CR"], errors="coerce").astype("Int64")
    citation_year_col = "CR"  # 引用年份列名
    cited_year_col = "PY"  # 替换为实际的被引用文献发表年份列名

    # 检查数据中是否有这些列
    if citation_year_col not in data.columns or cited_year_col not in data.columns:
        # 如果没有引用数据，我们可以尝试使用发表年份和计数来估计
        if "年份" in data.columns:  # 假设有年份列表示发表年份
            year_column = "年份"

            # 获取年份范围
            min_year = data[year_column].min()
            max_year = data[year_column].max()
            span = max_year - min_year

            # 统计每年的文献数量
            yearly_counts = data[year_column].value_counts().sort_index()
            years = yearly_counts.index.tolist()
            counts = yearly_counts.values

            # 计算文献年龄
            current_year = max(years) + 1  # 使用最新年份 + 1作为当前年份
            ages = [current_year - year for year in years]

            # 为了计算老化系数，我们需要归一化计数
            normalized_counts = counts / max(counts)

            # 对规范化计数取对数
            log_counts = np.log(normalized_counts)

            # 线性回归：log(C) = -a·t + b
            slope, intercept, r_value, p_value, std_err = stats.linregress(
                ages, log_counts
            )

            # 老化系数是线性回归的斜率的绝对值
            aging_coefficient = abs(slope)

            # 半衰期计算
            half_life = np.log(2) / aging_coefficient

            print(f"\n文献老化分析结果:")
            print(f"老化系数: {aging_coefficient:.4f}")
            print(f"半衰期: {half_life:.2f} 年")
            print(f"拟合优度(R²): {r_value**2:.4f}")

            # 可视化老化趋势
            plt.figure(figsize=(10, 6))
            plt.plot(ages, normalized_counts, "bo", alpha=0.7, label="实际数据")

            # 生成拟合曲线
            x_range = np.linspace(min(ages), max(ages), 100)
            y_fit = np.exp(intercept) * np.exp(slope * x_range)

            plt.plot(
                x_range,
                y_fit,
                "r-",
                label=f"拟合曲线 (a={aging_coefficient:.4f}, T₁/₂={half_life:.2f}年)",
            )
            plt.xlabel("文献年龄 (年)")
            plt.ylabel("归一化使用频率")
            plt.title("文献老化指数衰减模型")
            plt.yscale("log")  # 使用对数坐标更直观显示指数衰减
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

            # 可视化半衰期
            plt.figure(figsize=(10, 6))
            plt.plot(ages, normalized_counts, "bo-", alpha=0.7)
            plt.axhline(y=0.5, color="r", linestyle="--", label="半衰期阈值 (50%)")
            plt.axvline(
                x=half_life,
                color="g",
                linestyle="--",
                label=f"半衰期: {half_life:.2f}年",
            )
            plt.xlabel("文献年龄 (年)")
            plt.ylabel("归一化使用频率")
            plt.title("文献半衰期分析")
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

            # 分析不同学科领域的老化特征（如果有学科分类列）
            discipline_col = "学科"  # 替换为实际的学科列名
            if discipline_col in data.columns:
                disciplines = data[discipline_col].unique()

                plt.figure(figsize=(12, 8))
                for discipline in disciplines[:5]:  # 取前5个学科避免图表太乱
                    discipline_data = data[data[discipline_col] == discipline]
                    discipline_yearly = (
                        discipline_data[year_column].value_counts().sort_index()
                    )
                    discipline_years = discipline_yearly.index.tolist()
                    discipline_counts = discipline_yearly.values
                    discipline_ages = [current_year - year for year in discipline_years]

                    # 归一化
                    discipline_norm = discipline_counts / max(discipline_counts)

                    # 排序
                    sort_idx = np.argsort(discipline_ages)
                    discipline_ages_sorted = np.array(discipline_ages)[sort_idx]
                    discipline_norm_sorted = np.array(discipline_norm)[sort_idx]

                    plt.plot(
                        discipline_ages_sorted,
                        discipline_norm_sorted,
                        "o-",
                        label=discipline,
                    )

                plt.xlabel("文献年龄 (年)")
                plt.ylabel("归一化使用频率")
                plt.title("不同学科文献老化特征比较")
                plt.yscale("log")
                plt.legend()
                plt.grid(True, alpha=0.3)
                plt.show()
        else:
            print("未找到年份信息，无法分析文献老化。")
    else:
        # 如果有引用数据，我们可以直接计算引用年龄
        data["引用年龄"] = data[citation_year_col] - data[cited_year_col]

        # 计算每个年龄的引用次数
        citation_by_age = data["引用年龄"].value_counts().sort_index()
        ages = citation_by_age.index.tolist()
        citations = citation_by_age.values

        # 归一化引用次数
        normalized_citations = citations / max(citations)

        # 对规范化引用次数取对数
        log_citations = np.log(normalized_citations)

        # 线性回归：log(C) = -a·t + b
        slope, intercept, r_value, p_value, std_err = stats.linregress(
            ages, log_citations
        )

        # 老化系数是线性回归的斜率的绝对值
        aging_coefficient = abs(slope)

        # 半衰期计算
        half_life = np.log(2) / aging_coefficient

        print(f"\n文献老化分析结果:")
        print(f"老化系数: {aging_coefficient:.4f}")
        print(f"半衰期: {half_life:.2f} 年")
        print(f"拟合优度(R²): {r_value**2:.4f}")

        # 可视化老化趋势
        plt.figure(figsize=(10, 6))
        plt.plot(ages, normalized_citations, "bo", alpha=0.7, label="实际数据")

        # 生成拟合曲线
        x_range = np.linspace(min(ages), max(ages), 100)
        y_fit = np.exp(intercept) * np.exp(slope * x_range)

        plt.plot(
            x_range,
            y_fit,
            "r-",
            label=f"拟合曲线 (a={aging_coefficient:.4f}, T₁/₂={half_life:.2f}年)",
        )
        plt.xlabel("引用年龄 (年)")
        plt.ylabel("归一化引用频率")
        plt.title("文献老化指数衰减模型")
        plt.yscale("log")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()


# 运行文献老化分析
analyze_literature_aging(data)
```
